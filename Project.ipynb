```python
# Cognixia Project 
# Programmer: Senthooran Yogeswaran 
"""
Shortcuts to make reading easier to digest 
# Main Menu for Stocks Simulator: Will take you to the main menu code of the stock simulator
# # rps game menu : Will take you to the rock paper scissors menu code 
"""

#Variables for 
import matplotlib.pyplot as plt
import random
import csv


currentYear = [2017]
currentMonth  = [0]
endTime = 2018
#variables to identify stock names and months
stocks = []
monthList = ["Jan","Feb","Mar","April","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
#Dictionary array for stocks indicating the growth of each stock
stocksPrice = {}
#amount of money you start with in this simulator
capital = [5000]
#the variability in each stock can either be super volatile or safe
stocksVolatility = {}
#User's stock inventory 
stocksInventory = {}
#Events that can reduce stock value or increase it
potentialEvents = {"meteorstrike": -50, "trending": 20, "CEOstroke": -30, "Scandals": -15, "governmentfunding": 30}
currentEvent  = [""]
randStocks = [""] 


"""
Rock Paper scissors game that takes player input on valid choice and makes them face a round agaisnt a cpu.
Players are allowed to continue playing till desired  
"""
# rps game menu             
def rps_game():
    import random 

    options = ("rock", "paper", "scissors")
    player = None 
    computer = random.choice(options)


    playing = True 
    p1wins, p2wins = 0, 0
    cpu_wins = 0

    while playing:
            # We need to ensure the user picks a valid option: 
            while player not in options: 
                player = input("Enter a choice (rock, paper, or scissors): ")

            print("Rock... paper... scissors shoot! ")
            print(f"Player chose: {player}")
            print(f"Computer chose: {computer}")

            # Tie condition 
            if player == computer:
                print("It's a tie.")
            # User winning conditions 
            elif player == "rock" and computer == "scissors":
                p1wins += 1
                print("You win!")
            elif player == "paper" and computer == "rock":
                p1wins += 1
                print("You win!")
            elif player == "scissors" and computer == "paper":
                p1wins += 1
                print("You win!")
            else:
                cpu_wins += 1
                print("You lose!")

            if not input("Would you like to play? Y/N ").lower() == "y":
                playing = False

    print("Thank You, for playing! ")

    
"""
Python Simulator game that allows the user to experiment buying and selling various stocks amongst other features. 
"""
    
def menu_stocks(): 
    '''
    Prints instructions for the user in the menu to allow them to choose options.
    Also lets user know what the current month and year is
    '''
    print ("Welcome to the paper stock simulator! Your budget is $ " + str(capital[0]))
    print ("It is currently : " + monthList[currentMonth[0]] + " of " + str(currentYear[0]) + "\n")
    #prints current time period and balance
    print ("A: Buy a stock.")
    print ("B: Sell a stock.") 
    print ("C: Read news about stocks." ) 
    print ("D: See portfolio." ) 
    print ("E: See charts.")
    print ("F: End current month." )
    print ("Q: Quit. \n") 

	
def priceUpdate():
    '''
    Randomly updates the price for a stock depending on its volatility.
    '''  
    for i in range(6):
        effect = random.randint(0,1)
        stock = stocks[i]
        priceList = stocksPrice[stock]
        priceList.pop(0)
     #chooses between negative or positive effect, then pops the 0th place
    if(effect == 1):
        priceList.append(priceList[10] + ((random.randint(1,50)) * (stocksVolatility[stock])))
        stocksPrice[stock] = priceList
    # appends the new increased price of the stock, determined by random int multiplied by volatility
    elif(effect == 0):
        update = True
        while(update):
            decreased = (priceList[10] - ((random.randint(1,50)) * (stocksVolatility[stock]))) 
            if (decreased > 0):
                update = False
                priceList.append(decreased)
#appends the decreased price of the stock instead, but prevent the stock from going into negative by continuously generating new price until it is larger than 0.
        

    #updates the stock price randomly
    
    eventList = list(potentialEvents.keys())
    eventChoice = eventList[(random.randint(0,4))]
    currentEvent[0] = potentialEvents[eventChoice]
    #chooses random event
    randStocks[0] = stocks[(random.randint(0,5))]
    #chooses random stock
    priceList = list(stocksPrice[randStocks[0]])
    priceList[2] = priceList[2] + potentialEvents[eventChoice]  
    stocksPrice[randStocks[0]]



def purchase_stock():
    '''
  Allows user to purchase stocks from: the list of stocks 
  
    '''  
    #Commands prompts tels users current stock prices and volatility   
    Exist = True
    print("The current stocks and prices are: ")
    for p in range(5):
        priceList = stocksPrice[stocks[p]]
        print(stocks[p] + ": " + str(priceList[11]))
    print("The current volatility for each stock is the following: " + str(stocksVolatility)  + "\n")
    stock_input = input("Please choose only one stock to purchase from above.")

    while(Exist):
      # does that stock exist in the dictionary
        if (stock_input in stocksPrice):
          
            Exist = False
            priceList = stocksPrice[stock_input]
            price = priceList[11]
            print(price)
            shares_input = int(input("How many shares would you like to purchase?" + "\n"))
            bought = shares_input * price
            wallet = capital[0]
            # stocksPrice[stock_input] >= bought
            # Makes sure you have enough money to buy the amount of shares
            if(wallet > bought):
                balance = capital[0] - bought
                capital[0] = balance
                print ("Your budget now is: $" + str(capital[0]) + "\n")
                print ("You bought " + str(shares_input) + " share worth of " + str(stock_input) + "\n")
                stocksInventory[stock_input] += bought
                print(stocksInventory)  
            # error handling if budget is too low to buy stock purchase
            elif(wallet < bought):
                print("You don’t have enough money to make this stock purchase")
                return 
        # error handling for a stock that does not exist
        else:
            print("That stock doesn’t exist. Please choose a valid stock") 
            stock_input = input("Please choose only one stock to purchase from above.") 


def sell_stock():
    '''
  Allows user to sell stocks from their own portfolio:
  

    '''  
    #Commands prompts tels users current stock prices and volatility 
    Exist = True
    print("The current stocks and prices are: ")
    for p in range(5):
        priceList = stocksPrice[stocks[p]]
        print(stocks[p] + ": " + str(priceList[11]))
    print("The current volatility for each stock is the following: " + str(stocksVolatility)  + "\n") 
    stock_input_sell = input("Please choose a stock to sell.")
    #while the boolean is true 
    while(Exist):
      #check if the stock you want to sell exists 
        if (stock_input_sell in stocksPrice):
          #once the stock is found we set the boolean to false to exist the loop
            Exist = False
            #identify the stock from the dictionary and set it as a variable
            priceListsell = stocksPrice[stock_input_sell]
            #we want the last item in the dictionary 
            price_sell = priceListsell[11]
            #ask user how many stocks they would like to sell
            shares_input_sell = int(input("How many shares would you like to sell from this stock?"))
            #the amount you receive is equal to the amount of shares * how much the stock costs at the moment 
            sold = shares_input_sell * price_sell
            # we want create a sell limit variable to ensure you own that many of the stock 
            sell_limit = stocksInventory[stock_input_sell] - sold  
            #if the sell limit is greater than zero you add it to your balance and you are notified of your new budget 
            if(sell_limit >= 0):
                balance = capital[0] + sold
                capital[0] = balance 
                print("Your budget now is: $" + str(capital[0]))
                print("You sold " + str(shares_input_sell) + " shares worth of " + str(stock_input_sell))
                #remove the stocks you sold from your portfolio 
                stocksInventory[stock_input_sell] -= sold
            #if the sell limit is below zero; user is notified they cannot sell shares that they dont have
            elif(sell_limit < 0):
                print("You can't sell this stock because you don't own that many shares.")
        #if the stock does exist you are asked how many shares would you like to sell again
        else:
            print("The following stock doesn't exist to sell. Choose a valid one")
            shares_input_sell = int(input("How many shares would you like to sell from this stock"))
    return 

def stock_news():  
    """
    News function that changes the price of a random stock depending on dictionary potentialEvents.


    """
  # prints out the news for a random stock and updates the price
    print(str(currentEvent[0]) + "happened to " + str(randStocks[0]))
    print("Now the new stock price of " + str(randStocks[0]) + " is: " + str(stocksPrice[randStocks[0]]))

def portfolio_view():
    """
  Allows user to see progress with investments. Shows an overall portfolio of their stocks.
  Using list capital, the profit/loss can be determined. 

  """

    print("You currently have: $" + str(capital[0]))
    # if capital is greater than starting budget, user is in profit
    if ((capital[0] - 5000) > 0):
        print("This is " + str((capital[0] - 5000)) + "more than you started with, congrats! Keep up!")
  # if capital is less than starting budget, user is in a loss
    elif ((capital[0] - 5000) < 0):
        print("This is " + str((capital[0] - 5000)) + " less than you started with :( Keep working!")
  #Bar graph displaying stock portfolio user bought
    plt.barh(list(stocksInventory.keys()), list(stocksInventory.values()))
    plt.show()	
    
def end_current_month():
    """
    The user is allowed to end the current month and go to the next.
    This implies new stock prices will be updated 
    """
    # procede to the next month 
    currentMonth[0] += 1
    # if its the end of the year
    if currentMonth == 12:
    # progress the current year to 2018
        currentYear[0] += 1
    # set the month back to Jan
        currentMonth[0] = 0
    # quits if the endtime has reached
    if currentYear == endTime:
        quit()

def quit():
    """
    #displayes quitting statement
    """
    print("You have quit the paper stock simulator.")
    print("You managed to make $ " + str(capital[0]) + " with a portfolio of " + str(stocksInventory) +"\n")


def loadCSV(csv_input):
    """
    Converts the inputted csv file as "stocksVolatility", "stocksInventory", "stocks", "stocksPrice"
    """
    
    csv_handle = open(csv_input, "r")
    csv_reader = csv.DictReader(csv_handle)
    #load csv
    for next_row in csv_reader:
        stocksVolatility[next_row['Stocks']] = int(next_row['Volatility'])
        stocksInventory[next_row['Stocks']] = int(next_row['Inventory'])
        stocks.append(next_row['Stocks'])
        for i in range(12):
            stocksPrice.setdefault(next_row['Stocks'], []).append(int(next_row['Price' + str(i)]))
    
    csv_handle.close()
    return
    

# Main Menu for Python Simulator

def stocks_game():
    
    loadCSV("stocks.csv")

    menu = True

    while(menu):
        #menu instructions
        menu_stocks()
        #input 
        user_input = (input("Please choose an option from the menu: "))
        #allows user to purchase stocks
        if(user_input == "A" or user_input == "a"):
            purchase_stock()
         #allows user to sell stocks    
        elif(user_input == "B" or user_input == "b"):
            sell_stock()
         #allows user to see stock news 
        elif(user_input == "C" or user_input == "c"):
            stock_news()
         #allows user to see their portfolio 
        elif(user_input == "D" or user_input == "D"):
            portfolio_view()
         #allows user to end current month    
        elif(user_input == "F" or user_input == "f"):
            end_current_month()
          #updates all prices after you end the month 
            priceUpdate()
        elif(user_input == "Q" or user_input == "q"): 
            quit()
            menu = False
            #displayes quitting statement    
    
    
    
    
    

"""
Main menu method that prints out string files which resemble instructions 
"""
def main_menu():
    print("[1] Rock Paper Scissors Game. ")
    print("[2] Stock Simulator Game. ")
    print("[0] Exit. ")

main_menu() 
choice = int(input("Which game would you like to play? "))

# user is prompted to select an option
while choice != 0:
    if choice == 1:
        rps_game()
        #Play Rock Paper Scissors Game
    elif choice == 2:
        stocks_game()
        #Run Stock Simulator Game
    else:
        # Reiterate menu if invalid option is provided 
        print("Invalid option.")
    main_menu() 
    choice = int(input("Which game would you like to play? "))         
             






```
